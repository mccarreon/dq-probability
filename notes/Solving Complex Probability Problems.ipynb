# Complex Probability Problems

We still have not learned to tackle more complex probability problems, such as:
* What is the probability that it takes three flips or more for a coin to land heads up?
* What is the probability of a coin landing heads up 18 times in a row?
* Getting at least one 6 in four throws of a single six-sided die?
* Getting at least one double-six in 24 throws of two six-sided die?
* Getting 4 aces in a row when drawing cards from a 52-card deck?

The probability of event E happening can be described as so, provided they are mutually exclusive:
* P(E) = 1 - P(non-E)

### What is the probability that it takes three flips or more for a coin to land heads up

Let's tackle the first question, the probability of it taking three flips or more for a coin to land heads up. We'll call it event A. Event A and event non-A can be described as:
* A = {3,4,5,6,...,100,101,...}
* non-A = {1,2}

Event A contains infinite numbers because it's possible to never land on heads. We can use our formula from above to find P(A). 
* P(A) = 1 - P(non-A)

We will need to find P(non-A) first, since that is not an infinite amount of possibilities. Event A only happens if it takes three flips or more for a coin to land heads up. If a coin is flipped twice and heads comes up at least once, event non-A happens. 

The following outcomes are possible: 
* {HH,HT,TH,TT}

Thus, P(non-A) = 3/4 and P(A) = 1 - 3/4, P(A) = 1/4

What if this was extended to at least 4 flips? 
* Event non-B = {1,2,3}
* Outcomes: {HHH, HHT, HTT, TTT, TTH, THH, THT, HTH}

Thus, P(non-B) = 7/8 and P(B) = 1/8.

## Set Notation

In set theory, the opposite of a set is called its **complement**. Consider the event "getting a 2 when rolling a fair six-sided die" called event B. The set for event B and its opposite would be:
* B = {2}
* non-B = {1,3,4,5,6}, or B<sup>C</sup> = {1,3,4,5,6}

The equation we started with can be re-written like so:
* P(E $\cup$ E<sup>C</sup>) = P(E) + P(E<sup>C</sup>) = 1

Now let's consider the next question: the probability of flipping heads 18 times in a row. 

Let's start with the probability of getting heads twice in a row. The outcomes would look like this: 
* {HH,HT,TH,TT}
* P(HH) = 1/4

However, doing this 18 times in a row would lead to 262,144 possible outcomes. We can't realistically list out the outcomes. So, let's see a different approach. 

## Multiplication Rule of Probability 

P(HH) is the probability of a coin landing heads on the first flip (H<sub>1</sub>) **and** it lands heads up on the second flip (H<sub>2</sub>), so we want to find P(H<sub>1</sub> $\cap$ H<sub>2</sub>).

We can use the **multiplication rule of probability** to find P(H<sub>1</sub> $\cap$ H<sub>2</sub>) and multiply P(H<sub>1</sub>) and P(H<sub>2</sub>). 
* P(H<sub>1</sub> $\cap$ H<sub>2</sub>) = P(H<sub>1</sub>) * P(H<sub>2</sub>)
* P(H<sub>1</sub> $\cap$ H<sub>2</sub>) = 0.5 * 0.5 = 0.25

In general terms, the multiplication rule states that for two events *E<sub>1</sub>* and *E<sub>2</sub>*, the probability of both happening can be found by multiplying their probabilities together. In this form, the equation works only for events taht are **independent events**. This means that if E<sub>1</sub> happens, the probability of E<sub>2</sub> happening isn't affected. 

### What is the probability of a coin landing heads up 18 times in a row

So, coming back to our original question, we can find the possibility of heads being flipped 18 times in a row like so: 
* P(18 heads in a row) = 0.5<sup>18</sup>

Let's revisit the next two questions:
* Getting at least one 6 in four throws of a single six-sided die?
* Getting at least one double-six in 24 throws of two six-sided die?

### Getting at least one 6 in four throws of a single six-sided die

Let's focus on the first one for now, and refer to the event "getting at least one 6 in four throws of a single six-sided die" as event A. So, event A<sup>C</sup> would be not getting any 6 in four throws. Getting any of the outcomes A<sup>C</sup> = {1,2,3,4,5} four times in a row. So, a 5/6 chance four times in a row:
* P(A<sup>C</sup>) = (5/6)<sup>4</sup>

Now, we can get P(A):
* P(A) = 1 - (5/6)<sup>4</sup>

### Getting at least one double-six in 24 throws of two six-sided die

Let's refer to the event of rolling two sixes event A. The chance of rolling two sixes is 1/6 * 1/6, so P(A) = 1/36. This means that P(A<sup>C</sup>) = 1 - 1/36 = 35/36. Since we're throwing 24 times, the probability of P(A) not happening at all would be (35/36)<sup>24</sup>. Thus, P(A) = 1 - (35/36)<sup>24</sup>.

### Getting 4 aces in a row when drawing cards from a 52-card deck

Let's start with the probability of drawing an ace from a full deck:
* P(Ace) = 4/52
To determine whether or not P(AAAA) = (4/52)<sup>4</sup> depends on if we are replacing the cards back into the deck or not. If we don't put them back, it is not correct to use the multiplication rule as we have been using it. 

If we don't replace, the second draw probability is:
* P(Ace) = 3/51

So, for four aces in a row: 
* P(AAAA) = 4/52 * 3/51 * 2/50 * 1/49

**Sampling with replacement** would have us replace the cards after drawing and the formula would be:
* P(AAAA) = (4/52)<sup>4</sup>

**Sampling without replacement** would have us keep the cards out of the deck after drawing, so this would be the formula: 
* P(AAAA) = 4/52 * 3/51 * 2/50 * 1/49
